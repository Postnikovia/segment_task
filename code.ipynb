```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Импорт данных
train = pd.read_csv('contest_train.csv', delimiter = ',')
test = pd.read_csv('contest_test.csv', delimiter = ',')
# выделяем целевую величину
target = train['TARGET']
train.drop(['TARGET'], axis = 1, inplace =True)
#объеденяем выборки помечая их
train['istest'] = 0
test['istest'] = 1
data = pd.concat([test,train])
data
```

```python
print(target.value_counts())
# смотрим распределение целевой велечины
```

```python
lis =[]
for i in range(0,data.shape[1]):
    if data[data.columns[i]].count() < 0.7*len(data):
        lis.append(i)
lis = np.array(lis)
lis = data.columns[lis]
lis = lis.values
data.drop(lis, axis =1, inplace=True)

# Удаляем столбцы с большим количеством пропусков
```

```python
corr = data.corr()
corr
```

```python
#Записываем индексы признаков с высокой корреляцией
drop = []
for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        if corr[corr.columns[i]][j] >0.95 and corr[corr.columns[i]][j]!=1 :
            drop.append(i)
        if corr[corr.columns[i]][j] < -0.95:
            drop.append(i)

from itertools import groupby
#Удаляем дубликаты
drop = [el for el, _ in groupby(drop)]
#Получаем название столбцов которые нужно убрать
drop = np.array(drop)
drop = data.columns[drop]
drop = drop.values
#Удаляем фичи которые сильно скорелированы (1 из пары)
data.drop(drop,axis=1, inplace=True)
data
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators = 500, max_depth = 6, n_jobs = -1, class_weight = 'balanced')
data_feat = data.copy()
#Выделяем тренировочный тест для проведения кросс-валидации
train_f=data_feat[data_feat['istest']==0].copy()
train_f.drop(['istest','ID'],axis=1, inplace=True)
# Соединяем целевую величину и данные для грубой очистки миссингов
train_f['target'] = target.copy()
# Дропаем миссинги
train_f.dropna( inplace=True)
target_f = train_f['target']
train_f.drop(['target'],axis=1,inplace=True)
# train_test_split
delit = 0.8
index = int(delit*len(train_f))
feature_train = train_f[:index]
feature_test = train_f[index:]
feature_train_tar = target_f[:index]
feature_test_tar = target_f[index:]

forest.fit(feature_train,feature_train_tar)
from sklearn.metrics import  f1_score
from sklearn.metrics import accuracy_score
pred = forest.predict(feature_test)
# Посмотрим результат метрик для дальнейшего сравнения
print(f1_score(feature_test_tar, pred,average = 'weighted'))
print(accuracy_score(feature_test_tar, pred))       
# Выделяем фичи которые наиболее значимые для модели
importance = forest.feature_importances_
nazv = list(feature_train)

important = pd.DataFrame({'index':nazv,'importance':importance})
crit = important['importance'].max()*0.1
important = important[important['importance']>crit]
features = important['index']
len(features)
features
```

```python
#Оставляем  эти фичи  на наших данных
data2 = data[features].copy()
#Записываем метки разделения и id шники
data2['istest'] = data['istest']
data2['ID'] = data['ID']

data2
```

```python
#Работаем с миссингом
def rang_feach(i,data):
    val = data[data.columns[i]].value_counts().sort_values(ascending = False)
    return val.index[0]

for i in range(0, data2.shape[1]):
    razmer = data2[data2.columns[i]].value_counts().count() #Кол-во уникальных значений
    if razmer < 10:# Считаем что фича ранговая
        val = rang_feach(i,data2)
        data2[data2.columns[i]].fillna(val, inplace=True )
    else:#Cчитаем что фича номинальная
        data2[data2.columns[i]].fillna(data2[data2.columns[i]].mean(), inplace=True)
```

```python
#Возвращаем  трениовочные и тестовые  выборки
test = data2[data2['istest']==1].copy()
idi = test['ID']#Сейвим айдишники для ответов
test.drop(['istest','ID'],axis=1, inplace = True)
train = data2[data2['istest']==0].copy()
train.drop(['istest','ID'],axis=1, inplace = True)
train
```

```python
#Подбор параметров
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
forest = make_pipeline(RandomForestClassifier(n_jobs=-1, class_weight = 'balanced'))
param = [{'randomforestclassifier__n_estimators':[500,1000],
          'randomforestclassifier__max_depth':[10,12,13]}]
grid = GridSearchCV(forest,param, cv=5, scoring = 'f1_weighted')
grid.fit(train,target)
print('Лучшие параметры')
print(grid.best_params_)
print('Лучший результат')
print(grid.best_score_.mean())
```

```python
#кросс-валидация исходной модели
delit=0.9
index = int(delit*len(train))
ttrain = train[:index]
ttest = train[index:]
testtarget = target[index:]
traintarget = target[:index]

model = RandomForestClassifier(n_estimators=1000, max_depth = 12, n_jobs=-1,class_weight = 'balanced')

model.fit(ttrain,traintarget)
predtarget = model.predict(ttest)
print('Итоговая кроссвалидация')
print(f1_score(testtarget, predtarget,average = 'weighted'))
print(accuracy_score(testtarget,predtarget))
```

```python
#Проведем расчет по метрике lift
from sklearn.metrics import precision_score
def lift(testtarget,predtarget,target):
    nazv = [0,1,2]
    proc = []
    for i in pd.DataFrame(target.value_counts()).index:
        col = pd.DataFrame(target.value_counts()).iloc[i].values.copy()
        avg_proc = col/len(target)
        proc.append(avg_proc)
    stats = pd.DataFrame({'segment':nazv,'avg':proc})
    predict =  pd.DataFrame({'true':testtarget,'pred':predtarget})
    precision = []
    for  i in range(0,3):
        preds=predict[predict['pred']==i].copy()
        preds['itog'] = 0
        preds['itog'][preds['true']==preds['pred']]=1
        precis = preds['itog'].sum()/len(preds)
        precision.append(precis)
    stats['precision'] = pd.Series(precision).values
    stats['lift']=stats['precision']/stats['avg']
    return stats['lift']
# Тест того что все работает
print(lift(testtarget,predtarget,target))
```

```python
itog1 = []
itog2 = []
depth_list = []
n_estimators = []
for dept in [7,9,11,13,15]:
    for n_estimator in [100,1000]:
        model = RandomForestClassifier(n_estimators=n_estimator, max_depth = dept, n_jobs=-1,class_weight = 'balanced')
        model.fit(ttrain,traintarget)
        predtarget = model.predict_proba(ttest)
        predtarget = pd.DataFrame(predtarget)

        predtarget = predtarget.sort_values(by=[predtarget.columns[0]],ascending=False)
        idik = predtarget.index[:500].copy()
        clas = target[idik].copy()
        predict_clas_1 = clas.values
        itog1.append(lift(np.zeros(500),predict_clas_1,target)[0])

        predtarget = predtarget.sort_values(by=[predtarget.columns[1]],ascending=False)
        idik = predtarget.index[:200].copy()
        clas = target[idik].copy()
        predict_clas_2 = clas.values
        itog2.append(lift(np.ones(200),predict_clas_2,target)[1])
        depth_list.append(dept)
        n_estimators.append(n_estimator)

itog_data = pd.DataFrame({'lift1_segment':itog1,'lift2_segment':itog2,'depth':depth_list,'n_estimators':n_estimators})
itog_data
```
